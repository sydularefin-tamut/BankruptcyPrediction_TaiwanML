```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df=pd.read_csv("data.csv")
print(df.shape)
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
bankrupt= df.drop(" Net Income Flag", axis=1) 
X=bankrupt.drop("Bankrupt?", axis=1).values
y=bankrupt["Bankrupt?"].values
```

```python
X_Train, X_test, Y_Train, y_test = train_test_split(X,y,test_size=0.10,random_state=101)
```

```python
X_train, X_valid, y_train, y_valid= train_test_split(X_Train, Y_Train,test_size=0.25,random_state=101)
```

```python
X_train.shape, X_valid.shape, X_test.shape, y_train.shape, y_valid.shape,  y_test.shape
```

```python
import pickle

# Assuming you already have X_train, X_valid, X_test, y_train, y_valid, y_test defined

# Dumping the data to pickle files
with open('X_train.pkl', 'wb') as file:
    pickle.dump(X_train, file)

with open('X_valid.pkl', 'wb') as file:
    pickle.dump(X_valid, file)

with open('X_test.pkl', 'wb') as file:
    pickle.dump(X_test, file)

with open('y_train.pkl', 'wb') as file:
    pickle.dump(y_train, file)

with open('y_valid.pkl', 'wb') as file:
    pickle.dump(y_valid, file)

with open('y_test.pkl', 'wb') as file:
    pickle.dump(y_test, file)

# Loading the data from pickle files
with open('X_train.pkl', 'rb') as file:
    loaded_X_train = pickle.load(file)

with open('X_valid.pkl', 'rb') as file:
    loaded_X_valid = pickle.load(file)

with open('X_test.pkl', 'rb') as file:
    loaded_X_test = pickle.load(file)

with open('y_train.pkl', 'rb') as file:
    loaded_y_train = pickle.load(file)

with open('y_valid.pkl', 'rb') as file:
    loaded_y_valid = pickle.load(file)

with open('y_test.pkl', 'rb') as file:
    loaded_y_test = pickle.load(file)

# Checking the shapes to ensure data consistency
print(loaded_X_train.shape, loaded_X_valid.shape, loaded_X_test.shape, loaded_y_train.shape, loaded_y_valid.shape, loaded_y_test.shape)
```

```python
import pandas as pd

# Assuming you already have X_train, X_valid, X_test, y_train, y_valid, y_test defined as NumPy arrays

# Convert NumPy arrays to Pandas DataFrames
X_train_df = pd.DataFrame(X_train, columns=[f'feature_{i}' for i in range(X_train.shape[1])])
X_valid_df = pd.DataFrame(X_valid, columns=[f'feature_{i}' for i in range(X_valid.shape[1])])
X_test_df = pd.DataFrame(X_test, columns=[f'feature_{i}' for i in range(X_test.shape[1])])

y_train_df = pd.DataFrame(y_train, columns=['target'])
y_valid_df = pd.DataFrame(y_valid, columns=['target'])
y_test_df = pd.DataFrame(y_test, columns=['target'])

# Concatenating features and labels for training data
training_data = pd.concat([X_train_df, y_train_df], axis=1)

# Concatenating features and labels for test data
test_data = pd.concat([X_test_df, y_test_df], axis=1)

# Saving dataframes to CSV files
training_data.to_csv('train.csv', index=False)
test_data.to_csv('test.csv', index=False)
```
